# 5. Изучите документацию pandas в части описания методов merge и join.

```python
np.random.seed(0)
data1 = {
    'key': np.random.choice(['A', 'B', 'C'], 10),
    'value1': np.random.randint(1, 100, 10)
}
data2 = {
    'key': np.random.choice(['B', 'C', 'D'], 10),
    'value2': np.random.randint(1, 100, 10)
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print("Первый набор данных:")
print(df1)
print("\nВторой набор данных:")
print(df2)
```

```python
merge_inner = pd.merge(df1, df2, on='key', how='inner')
print("\nРезультат внутреннего соединения (merge):")
print(merge_inner)
```

```python
merge_outer = pd.merge(df1, df2, on='key', how='outer')
print("\nРезультат внешнего соединения (merge):")
print(merge_outer)
```

```python
merge_left = pd.merge(df1, df2, on='key', how='left')
print("\nРезультат левого соединения (merge):")
print(merge_left)
```

```python
merge_right = pd.merge(df1, df2, on='key', how='right')
print("\nРезультат правого соединения (merge):")
print(merge_right)
```

```python
join_outer = df1.join(df2.set_index('key'), on='key', how='outer', lsuffix='_left', rsuffix='_right')
print("\nРезультат внешнего соединения (join):")
print(join_outer)
```

# 6. Познакомьтесь с форматом представления дат strftime.

Ниже приведены некоторые основные символы формата, используемые в методе strftime():

%Y: Год в четырехзначном формате (например, 2024). <p>
%y: Год в двузначном формате (например, 24 для 2024). <p>
%m: Месяц в числовом формате (01 для января, 02 для февраля и т.д.). <p>
%d: День месяца (01-31). <p>
%H: Часы в 24-часовом формате (00-23). <p>
%I: Часы в 12-часовом формате (01-12). <p>
%M: Минуты (00-59). <p>
%S: Секунды (00-59). <p>
%A: Полное название дня недели (например, Monday). <p>
%a: Сокращенное название дня недели (например, Mon). <p>
%B: Полное название месяца (например, January). <p>
%b: Сокращенное название месяца (например, Jan). <p>
%p: AM или PM.
