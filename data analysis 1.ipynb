```python
#2
import numpy as np

def matrix(N, el1, d):
    progression = np.arange(el1, el1 + N * d, d)
    matrix = np.zeros((N, N))
    np.fill_diagonal(matrix, progression)
    return matrix

N = 7
el1 = 2.1
d = 3.7

itog_matrix = matrix(N, el1, d)
print(itog_matrix)
```

```python
#9
import pandas as pd
df.drop(columns=["Id", "DistrictId"], inplace=True)

new_titles = {
    'Rooms': 'Количество комнат',
    'Square': 'Площадь',
    'LifeSquare': 'Жилая площадь',
    'KitchenSquare': 'Площадь кухни',
    'Floor': 'Этаж',
    'HouseFloor': 'Этажность дома',
    'HouseYear': 'Год постройки дома',
    'Ecology_1': 'Экология 1',
    'Ecology_2': 'Экология 2',
    'Ecology_3': 'Экология 3',
    'Social_1': 'Социальный показатель 1',
    'Social_2': 'Социальный показатель 2',
    'Social_3': 'Социальный показатель 3',
    'Healthcare_1': 'Здравоохранение 1',
    'Helthcare_2': 'Здравоохранение 2',
    'Shops_1': 'Магазины 1',
    'Shops_2': 'Магазины 2',
    'Price': 'Цена'
}

df.rename(columns=new_titles, inplace=True)

df
```

```python
#11
last10 = df.tail(10).copy()
last10.drop(columns='Цена', inplace=True)
mean_price = df['Цена'].mean()
last10['Цена'] = mean_price
new_df = pd.concat([df.iloc[:-10], last10])
new_df
```

```python
#12
last5 = df.iloc[:, -5:]

mean_price = df['Цена'].mean()
last5 = last5[df['Цена'] >= mean_price]

new_df = pd.concat([df.iloc[:, :-5], last5], axis=1)
new_df
```

```python
#13
result = df.groupby('Floor').agg({'Price': 'mean', 'Rooms': 'count'})
result.rename(columns={'Цена': 'Средняя цена', 'Количество комнат': 'Среднее Количество комнат'}, inplace=True)
result
```

```python
#14
writer = pd.ExcelWriter('data1.xlsx')
result.to_excel(writer, sheet_name='Данные')
writer.save()
```

```python
#17
import matplotlib.pyplot as plt
data = pd.read_csv('Data.csv')
room_counts = data['Rooms'].value_counts()

max_rooms = room_counts.idxmax()
explode = [0.1 if rooms == max_rooms else 0 for rooms in room_counts.index]

plt.figure(figsize=(8, 8))
plt.pie(room_counts, labels=room_counts.index, autopct='%1.1f%%', startangle=90, explode=explode, colors = ['red', 'blue', 'orange', 'pink', 'yellow', 'black', 'green'])
plt.title('Распределение квартир по количеству комнат')
plt.axis('equal') 
plt.show()
```

```python
#19
data = pd.read_csv('data.csv')

plt.figure(figsize=(16, 12))

plt.subplot(2, 2, 1)
sns.scatterplot(x='Rooms', y='Price', data=data)
plt.title('Rooms / Price')

plt.subplot(2, 2, 2)
sns.scatterplot(x='Square', y='Price', data=data)
plt.title('Square / Price')

plt.subplot(2, 2, 3)
sns.scatterplot(x='HouseFloor', y='Price', data=data)
plt.title('HouseFloor / Price')

plt.subplot(2, 2, 4)
sns.scatterplot(x='HouseYear', y='Price', data=data)
plt.title('HouseYear / Price')

plt.tight_layout()
plt.show()
```

```python
#20 
data = pd.read_csv('data.csv')

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.kdeplot(data['Price'], shade=True)
plt.title('Ядерная оценка плотности для Price')

plt.subplot(1, 2, 2)
sns.kdeplot(x=data['Price'], y=data['HouseFloor'], cmap='viridis', fill=True, levels=50)
plt.title('Двумерная KDE для Price и HouseFloor')

plt.tight_layout()
plt.show()

```

```python
#21
data = pd.read_csv('Data.csv')

plt.figure(figsize=(8, 6))
sns.boxplot(y=data['Square'])
plt.title('Ящиковая диаграмма для Square')

plt.show()
```

```python
#22
features = ['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']
data = data[features]
g = sns.PairGrid(data)
g.map_diag(sns.histplot, kde_kws={'color': 'k'})
g.map_lower(sns.kdeplot, cmap='Blues_d')
g.map_upper(sns.scatterplot)
g.fig.suptitle('Попарные отношения признаков Rooms, Square, HouseFloor, Price', y=1.02)
plt.show()
```

```python
#23
features = ['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']
data = data[features]
corr_matrix = data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='aqua', fmt=".2f", annot_kws={"size": 10})
plt.title('Тепловая карта матрицы корреляции')
plt.show()
```
