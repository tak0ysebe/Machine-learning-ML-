```python
#4
failed_students = df["Балл"][2:] < 40
failed_students = failed_students.sum() / len(df) * 100
failed_students
```

```python
#5
passed_students = df["Балл"][2:] >= 40
passed_students = passed_students.sum() / len(df) * 100
passed_students
```

```python
labels = ["Сдавшие экзамен", "Не сдавшие экзамен"]
values = [passed_students, failed_students]
plt.pie(values, labels=labels, autopct='%1.1f%%')#!!!!!!!!!!!!!!!!!!!!!!!!!!!
plt.axis("equal")
plt.show()
```

```python
#6
df["Балл"][2:].plot(kind="density")
plt.xlabel("Баллы за экзамен")
plt.ylabel("Плотность")
plt.title("Ядерная оценка плотности распределения баллов за экзамен")
plt.show()
```

```python
#7
excellent_students = sum(df["Балл"][2:] > 85)
good_students = sum((df["Балл"][2:] > 69) & (df["Балл"][2:] < 86)) 
satisfactory_students = sum((df["Балл"][2:] < 49) & (df["Балл"][2:] < 68))
unsatisfactory_students = sum(df["Балл"][2:] < 50) 
labels = ["Отлично", "Хорошо", "Удовлетворительно", "Неудовлетворительно"]
values = [excellent_students, good_students, satisfactory_students, unsatisfactory_students]

plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.axis("equal")
plt.title("Процентное соотношение оценок за экзамен")
plt.show()
```

```python
#12
result = {}
for x in df["Задания с развёрнутым ответом"][2:-1]:
    k = 0
    x = x.replace(')', '(')
    x = x.split('(')
    x = x[:-1]
    for i in range(0, len(x), 2):
        k += 1
        if x[i] in '1234':
            result[k] = result.get(k, 0) + 1

print(dict(sorted(result.items())))
```

```python
#13
#i
df148 = df[df['№ школы']==148]
result = {}
for x in df148["Задания с кратким ответом"][2:-1]:
    k = 0
    for i in x:
        k += 1
        if i == '+':
            result[k] = result.get(k, 0) + 1

labels = ['B' + str(key) for key, value in sorted(result.items())]
values = [value for key, value in sorted(result.items())]
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.axis("equal")
plt.title('Процентное соотношение оценок за задания с кратким ответом в 148 школе')
plt.show()
```
