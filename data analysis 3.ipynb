```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Функция для генерации случайной даты в заданном диапазоне
def random_date(start_date, end_date):
    return start_date + timedelta(seconds=random.randint(0, int((end_date - start_date).total_seconds())))

# Создание первого датасета
def create_dataset1(num_entries):
    data = {
        'Id': ['ID_' + str(i) for i in range(1, num_entries + 1)],
        'Channel': [random.choice(['Email', 'Phone', 'Chat']) for _ in range(num_entries)],
        'category': [random.choice(['Technical', 'Billing', 'General']) for _ in range(num_entries)],
        'Sub-category': [random.choice(['Hardware', 'Software', 'Payment Issue']) for _ in range(num_entries)],
        'Issue_reported_Date': [datetime(2023, random.randint(1, 12), random.randint(1, 28)) for _ in range(num_entries)],
        'Issue_responded_Date': [random_date(datetime(2024, 1, 1), datetime.now()) for _ in range(num_entries)],
        'Survey_response_Date': [random_date(datetime(2024, 1, 1), datetime.now()) for _ in range(num_entries)],
        'Agent_name': ['Agent_' + str(random.randint(1, 10)) for _ in range(num_entries)],
        'CSAT Score': [random.randint(1, 5) for _ in range(num_entries)]
    }
    return pd.DataFrame(data)

# Создание второго датасета
def create_dataset2(num_entries):
    data = {
        'Id': ['ID_' + str(i) for i in range(1, num_entries + 1)],
        'Channel': [random.choice(['Email', 'Phone', 'Chat']) for _ in range(num_entries)],
        'category': [random.choice(['Technical', 'Billing', 'General']) for _ in range(num_entries)],
        'Sub-category': [random.choice(['Hardware', 'Software', 'Payment Issue']) for _ in range(num_entries)],
        'Issue_reported_Date': [datetime(2023, random.randint(1, 12), random.randint(1, 28)) for _ in range(num_entries)],
        'Issue_responded_Date': [random_date(datetime(2024, 1, 1), datetime.now()) for _ in range(num_entries)],
        'Survey_response_Date': [random_date(datetime(2024, 1, 1), datetime.now()) for _ in range(num_entries)],
        'Agent_name': ['Agent_' + str(random.randint(1, 10)) for _ in range(num_entries)],
        'CSAT Score': [random.randint(1, 5) for _ in range(num_entries)]
    }
    return pd.DataFrame(data)

# Создание двух датасетов
dataset1 = create_dataset1(100)
dataset2 = create_dataset2(100)
```

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

# Создание образцовых данных
combined_data_example = pd.concat([dataset1, dataset2], ignore_index=True)

# Создание subplot'ов для категориальных данных
fig, axs = plt.subplots(3, 2, figsize=(20, 15))

# Столбчатая диаграмма для Channel
combined_data_example['Channel'].value_counts().plot(kind='bar', ax=axs[0, 0], color='orange')
axs[0, 0].set_title('Распределение по каналам связи')
axs[0, 0].set_xlabel('Канал')
axs[0, 0].set_ylabel('Количество')

# Столбчатая диаграмма для category
combined_data_example['category'].value_counts().plot(kind='bar', ax=axs[0, 1], color='red')
axs[0, 1].set_title('Распределение по категориям')
axs[0, 1].set_xlabel('Категория')
axs[0, 1].set_ylabel('Количество')

# Столбчатая диаграмма для Sub-category
combined_data_example['Sub-category'].value_counts().plot(kind='bar', ax=axs[1, 0], color='green')
axs[1, 0].set_title('Распределение по подкатегориям')
axs[1, 0].set_xlabel('Подкатегория')
axs[1, 0].set_ylabel('Количество')

# Визуализация числовых данных (CSAT Score)
axs[1, 1].hist(combined_data_example['CSAT Score'], bins=5, color='pink', edgecolor='black')
axs[1, 1].set_title('Распределение оценок удовлетворенности клиентов (CSAT)')
axs[1, 1].set_xlabel('Оценка')
axs[1, 1].set_ylabel('Частота')

# Визуализация остальных числовых данных
numerical_columns = ['Id', 'Issue_reported_Date', 'Issue_responded_Date', 'Survey_response_Date']
for i, col in enumerate(numerical_columns):
    plt.figure(figsize=(8, 6))
    plt.hist(combined_data_example[col], bins=5, color='blue', edgecolor='black')
    plt.title(col)
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.show()

plt.tight_layout()
plt.show()
```
